```python
import pandas as pd

df = pd.read_csv('borg_traces_data.csv')

df.head()
```

```python
df.info()
```

```python
failed_events = ['FAIL','EVICT','LOST','KILL']
df.loc[df['event'].isin(failed_events),'failed'] = 1

df.head(20)
```

```python
null_idx = {}

for idx,val in enumerate(df.isnull().sum()):
    if val>0:
        null_idx[idx] = val

print(null_idx)
df.drop(df.columns[26:28],inplace=True,axis=1)
df.info()
```

```python
df.drop(['Unnamed: 0','user','collection_name','collection_logical_name'],inplace=True,axis=1)

df.info()
```

```python
df.drop(['machine_id'],inplace=True,axis=1)
```

```python
df.info()
```

```python
resources = df[df.columns[8]].dropna()

cpus = []
memory = []
```

```python
import ast

for resource in resources:
    resource = ast.literal_eval(resource)
    cpus.append(resource['cpus'])
    memory.append(resource['memory'])
```

```python
print(resources[0])
```

```python
fill_na = {'cpus':sum(cpus)/len(cpus),'memory':sum(memory)/len(memory)}
print(fill_na)
```

```python
df.loc[df['resource_request'].isnull(),'resource_request'] = fill_na
```

```python
null_idx = df[df['resource_request'].isnull()].index.tolist()
df.loc[null_idx,'resource_request'] = str(fill_na)
```

```python
import math
scaling = df['vertical_scaling'].dropna().tolist()

avg_scaling = sum(scaling)/len(scaling)
print(set(scaling))
avg_scaling = math.ceil(avg_scaling)
```

```python
df.loc[df['vertical_scaling'].isna(),'vertical_scaling'] = avg_scaling

df.isnull().sum()
```

```python
scheduler = df['scheduler'].dropna().tolist()
avg_scheduler = sum(scheduler)/len(scheduler)
print(avg_scheduler)
avg_scheduler = 0

df.loc[df['scheduler'].isna(),'scheduler'] = avg_scheduler
df.isnull().sum()
```

```python
df.drop(['cluster','event','tail_cpu_usage_distribution','cpu_usage_distribution'],inplace=True,axis=1)

df.info()
```

```python
obj_idx = df.select_dtypes(include=['object']).columns.tolist()

clean = df.drop(obj_idx,axis=1)
clean.info()
```

```python
exec_time = df['end_time'] - df['start_time']
```

```python
df.drop(['start_time','end_time'],inplace=True,axis=1)
df['exec_time'] = exec_time
df.head()
```

```python
df.drop(['constraint','random_sample_usage'],inplace=True,axis=1)
df.info()
```

```python
resources = df['resource_request'].tolist()
cpus.clear()
memory.clear()

for resource in resources:
    resource = ast.literal_eval(resource)

    cpus.append(resource['cpus'])
    memory.append(resource['memory'])

df.drop(['resource_request'],inplace=True,axis=1)
```

```python
df['req_cpus'] = cpus
df['req_memory'] = memory

df.head()
```

```python
cpus.clear()
memory.clear()
usage = df['average_usage'].tolist()

for resource in usage:
    resource = ast.literal_eval(resource)

    cpus.append(resource['cpus'])
    memory.append(resource['memory'])
df.drop(['average_usage'],axis=1,inplace=True)
df['avg_cpus'] = cpus
df['avg_memory'] = memory
```

```python
cpus.clear()
memory.clear()
usage = df['maximum_usage'].tolist()
null_idx.clear()

for idx,resource in enumerate(usage):
    resource = ast.literal_eval(resource)
    if not resource:
        null_idx.append(idx)
        continue
    
    cpus.append(resource['cpus'])
    memory.append(resource['memory'])
```

```python
df.drop(['maximum_usage'],inplace=True,axis=1)
```

```python
df['max_cpus'] = cpus
df['max_memory'] = memory
```

```python
df.head()
```

```python
df.drop(['start_after_collection_ids'],inplace=True,axis=1)
df.info()
```

```python
df.drop(['collections_events_type'],inplace=True,axis=1)
df.info()
```

```python
df.to_csv('regression_data.csv')
```
