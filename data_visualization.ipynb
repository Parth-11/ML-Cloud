```python
import pandas as pd

df = pd.read_csv('borg_traces_data.csv')

df.info()
```

```python
null_labels = {}
for label,i in enumerate(df.isnull().sum()):
    if i !=0:
        null_labels[label] = i

print(null_labels)
    
```

```python
labels = [df.columns[26],df.columns[27]]
```

```python
df.drop(labels,axis=1,inplace=True)
df = df.dropna()
```

```python
resources = (df[df.columns[10]].dropna())
```

```python
import ast
cpus = []
memorys = []
lst = []

for resouce in resources:
    resouce = ast.literal_eval(resouce)
    cpus.append(resouce['cpus'])
    memorys.append(resouce['memory'])
```

```python
df = df.dropna()
df.info()
```

```python
df.drop(df.columns[10],inplace=True,axis=1)
df['cpus'] = cpus
df['memory'] = memorys
```
