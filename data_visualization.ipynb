```python
failed_events = ['FAIL','EVICT','LOST','KILL']

df.loc[df['event'].isin(failed_events),'failed'] = 1
df.drop(['event'],inplace=True,axis=1)
df.head()
```

```python
obj = df.corr()

obj['failed']
```

```python
avg_cpus = []
avg_memorys = []
for elem in avg_usg:
    elem = ast.literal_eval(elem)
    avg_cpus.append(elem['cpus'])
    avg_memorys.append(elem['memory'])
```

```python
max_cpus = []
max_memory = []

for elem in max_usg:
    elem = ast.literal_eval(elem)
    if not elem:
        continue
    max_cpus.append(elem['cpus'])
    max_memory.append(elem['memory'])
```

```python
df['avg_cpus'] = avg_cpus
df['avg_memory'] = avg_memorys
df['max_cpus'] = max_cpus
df['max_memory'] = max_memory
df.head()
```

```python
obj = df.corr()
drop_col = []
for idx,val in enumerate(obj['failed']):
    if abs(val)<=0.02:
        drop_col.append(idx)

print(drop_col)
```

```python
drop_col = [df.columns[i] for i in drop_col]
drop_col.remove('vertical_scaling')
print(drop_col)
```

```python
df.drop(drop_col,inplace=True,axis=1)
df.head()
```

```python
df.to_csv("final_dataset.csv")
```
